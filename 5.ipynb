```python
import numpy as np

# Definimos el vector de medias para el primer grupo de datos
mu1 = [0, 0]

# Definimos la matriz de covarianza para el primer grupo de datos
sigma1 = np.array([[1, 0.5], [0.5, 1]])

# Definimos la misma matriz de covarianza para el segundo grupo de datos
sigma2 = np.array([[1, 0.5], [0.5, 1]])

# Generamos datos aleatorios para el primer grupo de datos siguiendo una distribucin normal multivariada
X = np.random.multivariate_normal(mu1, sigma1, 5000)

# Aadimos una traslacin a los datos del segundo grupo para que estn separados
X[2000:, 0] = X[2000:, 0] + 2
X[2000:, 1] = X[2000:, 1] + 2

# Creamos un vector de etiquetas para los datos, donde 0 representa el primer grupo y 1 el segundo
y = np.ones((5000,))
y[:2000] = 0

# Definimos una funcin simple para demostrar su uso en un lambda
def c(x):
    a = x ** 2
    return a

# Ejemplo de aplicacin de la funcin c
c(4)

# Visualizamos los datos generados
import matplotlib.pyplot as plt

# Creamos un grfico de dispersin para visualizar los datos
fig, ax = plt.subplots(1, 1, figsize=(16, 19))
plt.scatter(X[:, 0], X[:, 1], c=list(map(lambda x: "blue" if x == 0 else "red", y)))
plt.show()
```
